```python
filenames=["1.0Hz_5VPP_delay100.dat",
"1112.0Hz_5VPP_delay100.dat",
"2223.0Hz_5VPP_delay100.dat",
"3334.0Hz_5VPP_delay100.dat",
"4445.0Hz_5VPP_delay100.dat",
"5556.0Hz_5VPP_delay100.dat",
"6667.0Hz_5VPP_delay100.dat",
"7778.0Hz_5VPP_delay100.dat",
"8889.0Hz_5VPP_delay100.dat"]
```

```python
nfreq=len(filenames)
```

```python
nfreq=len(filenames)
data=[]
for i in range(nfreq):
    print("open: ",filenames[i])
    dumm=np.loadtxt(folder+filenames[i],delimiter=",")
    data.append(dumm)
```

```python
plt.plot(data[0],'ko',linestyle='solid')
```

```python
compfreq=[]
for i in range(nfreq):
    y=data[i]
    fig=plt.figure(figsize=(15,5))
    plt.subplot(121)
    plt.plot(y)
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.show()
```
